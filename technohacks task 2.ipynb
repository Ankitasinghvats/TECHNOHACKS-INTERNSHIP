## Finding mean medain mode standard deviation of Test data (option 1)

```python
data.describe()
```

```python

# deal with null values
data['Age'].fillna(data['Age'].median(), inplace=True)
data.dropna(subset=['Fare'], inplace=True)
data.drop(columns=['Cabin'], inplace=True)
data.isnull().sum()
     
     
```

```python
data.info()
```

```python
numeric_columns = data.select_dtypes(include=[int, float])
```

```python
numeric_columns
```

## Finding mean medain mode standard deviation of Test data (option 2)

```python
mean_values = numeric_columns.mean()
print("The mean values for each numeric column are:")
print(mean_values)

median_values = numeric_columns.median()
print("\nThe median values for each numeric column are:")
print(median_values)

mode_values = numeric_columns.apply(lambda x: stats.mode(x)[0][0])
print("\nThe mode values for each numeric column are:")
print(mode_values)

std_values = numeric_columns.std()
print("\nThe standard deviation values for each numeric column are:")
print(std_values)
```
